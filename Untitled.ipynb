```python
#To plot sin and cos wave
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# to define independent variable abd dependent variable

t = np.arange(0,1,0.001)
xt = np.sin(2*np.pi*3*t)
```

```python
# to plot the signal

plt.plot(t,xt)
plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
# How to plot sampled sin and cos signal
```

```python
n = np.arange(0,1, 0.001)
xn = np.sin(2*np.pi*3*n)
```

```python
plt.stem(n,xn)
plt.title('Sampled sin wave')
plt.xlabel('------>n')
plt.ylabel('Amplitude')
```
