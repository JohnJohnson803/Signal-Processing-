```python
import numpy as np

def myFunction(t):
    return t/ (2*np.pi)

def trapezoid_integration(func, a, b, N):
    deltaX = (b - a) / N
    output = func(a) + func(b)
    for i in range(1, N):  # Corrected the range to include all subintervals
        output += 2 * func(a + deltaX * i)
    return output * (deltaX / 2)

def integrand_an(func, t, n, omega):
    return func(t) * np.sin(n * omega * t)

def compute_an(func, a, b, N, omega, T):
    def integrand(t):
        return integrand_an(func, t, n, omega)
    integral = trapezoid_integration(integrand, a, b, N)
    return integral * (1 / T)

an = []

a = 0
T = 2 * np.pi
b = T
omega = 2 * np.pi / T
number_of_sub_intervals = 100

for n in range(1, 100):
    an.append(compute_an(myFunction, a, b, number_of_sub_intervals, omega, T))

print(an)

```
