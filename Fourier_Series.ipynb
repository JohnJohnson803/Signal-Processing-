```python
import numpy as np
import matplotlib.pyplot as plt
T = 1
w = 2*np.pi / T

f = []
ts = []

t = 0
dt = 0.25

while t<=T:
    ftemp = 0.5*t
    f.append(ftemp)
    ts.append(t)
    t = dt + t
print(f)
print(ts)
```

```python
xpoints = np.array(ts)
ypoints = np.array(f)
plt.plot(xpoints, ypoints)
plt.xlabel("t [s]")
plt.ylabel("f(s)")
plt.show()

```

```python
xpoints = np.array(ts)
ypoints = np.array(f)
plt.plot(xpoints + T, ypoints)
plt.xlabel("t [s]")
plt.ylabel("f(s)")
plt.show()
```
